```python
def test4perfect(number):
    return number == sum([d for d in range(1, number) if number % d == 0])


perfectnums = []
for number in range(2, 1001):
    if test4perfect(number):
        perfectnums.append(number)
print(perfectnums)
```

```python
numberarrange = np.arange(2, 51)

a = []
b = []
c = []

for m in numberarrange:
    for n in range(numberarrange[0], m):
        if m > n:
            a.append(m ** 2 - n ** 2)
            b.append(2 * m * n)
            c.append(m ** 2 + n ** 2)
fig, ax = plt.subplots(1, figsize=(10, 7))
plt.scatter(a, b, c=c, marker='o', alpha=.5, linewidths=.5, edgecolors='k')
plt.axis('off')
plt.show()

```

```python
n = 100
matr = np.zeros((n, n))
matr1 = np.zeros((n, n))
matr3 = np.zeros((n, n))
matr4 = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        matr[i, j] = np.sqrt(i ** 2 + j ** 2)
        matr3[i, j] = np.cbrt(i ** 3 + j ** 3)
        if matr[i, j] % 1:
            matr1[i, j] = 1
        if matr3[i, j] % 1:
            matr4[i, j] = 1
plt.subplot(221)
plt.imshow(matr)
plt.subplot(222)
plt.imshow(matr1)
plt.subplot(223)
plt.imshow(matr3)
plt.subplot(224)
plt.imshow(matr4)
plt.show()
```
