```python
import sympy as sym
from IPython.display import display
from IPython.core.display import Math
import numpy as np
import matplotlib.pyplot as plt
from sympy.abc import x

sym.init_printing()
```

```python
y = x**2-3*x


for i in range(-10,12):
    plt.plot(i,y.subs(x,i),'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %y)
    
plt.show()
```

```python
y = sym.sqrt(abs(x))
xList = list(range(-20, 21))

for xI in xList:
    plt.plot([0,  xI], [0, y.subs(x, xI)],linewidth=3)

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot([0, 2], [0, 0],linewidth=3,color='g')
plt.plot([2, 2], [0, 2],linewidth=3,color='r')
plt.plot([2, 0], [2, 2],linewidth=3,color='k')
plt.plot([0, 0], [2, 0],linewidth=3,color='y')
plt.axis('square')
plt.axis([-2.5,4.5,-3,5])

plt.show()
```

```python
x = [-10, 10]
m = [0.7, -1.25]
b = [-2, 0.75]


for i in range(len(m)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x, y, label='y=%sx+%s' % (m[i], b[i]))

plt.axis('square')
plt.grid()
plt.legend()

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphik')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
axis = plt.gca()
plt.plot([0, 0], axis.get_xlim(), ('k--'))
plt.plot(axis.get_ylim(), [0, 0], ('k--'))

plt.show()
```
