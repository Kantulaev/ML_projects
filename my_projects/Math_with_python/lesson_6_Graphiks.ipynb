```python
import sympy as sym
from IPython.display import display
from IPython.core.display import Math
import numpy as np
import matplotlib.pyplot as plt
from sympy.abc import x
from matplotlib.patches import Polygon
import sympy.plotting.plot as symplot
import scipy as sp
from scipy.signal import find_peaks

sym.init_printing()
```

```python
from cProfile import label


a = 1
k = 10
h = -10
n = 100
p = 1/(4*a)

x = np.linspace(h+1, h-1, 100)
y = a*(x-h)**2 + k

plt.plot(x, y, label='Parabola')
plt.plot(h, k, 'ro', label='Vertex')
plt.plot(h, k+p, 'co', label='Focus')
plt.plot(x[[0, -1]], [k-p, k-p], 'y', label='Directrix')

plt.legend()
plt.grid()

plt.show()
```

```python
x = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, x)
s = 2

Z = np.exp(-(X**2+Y**2)/s)

plt.axis('off')
plt.imshow(Z)
```

```python
h = [-1.5, 1.5]

x = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, x)
for r in np.linspace(.5, 3, 150):
    for hi in h:
        Fxy = Y**2 + (X-hi)**2 - r**2
        plt.contour(X, Y, Fxy, 0, colors=[r/3, r/3, r/3])

plt.plot(h, [0, 0], 'k', linewidth=3)

plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```

```python
n = 16
k = np.linspace(-4, 4, n)
b = 4
a = abs(k)


x = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, x)
for i in range(0, n):
    Fxy = (Y-k[i])**2/b**2 + X**2/a[i]**2 - 1
    plt.contour(X, Y, Fxy,0,colors=[i/n,0,i/n])

plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```

```python
n = 60
ab = np.linspace(1,5, n)

x = np.linspace(-8, 8, 200)
X, Y = np.meshgrid(x, x)
for i in range(0, n):
    Fxy = X**2/ab[i]**2 - Y**2/ab[i]**2 - 1
    plt.contour(Fxy, 0, colors=[i/n, 0, i/n])
    Fxy1 = Y**2/ab[i]**2 - X**2/ab[i]**2 - 1
    plt.contour(Fxy1, 0, colors=[0, i/n, 0])

plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```
