```python
import sympy as sym
from IPython.display import display
from IPython.core.display import Math
import numpy as np
import matplotlib.pyplot as plt
from sympy.abc import x
from matplotlib.patches import Polygon
import sympy.plotting.plot as symplot
import scipy as sp
from scipy.signal import find_peaks

sym.init_printing()
```

```python
t = np.linspace(-20,20,1000)
x = t * np.cos(t)**3
y = t * np.sin(t)**3

plt.plot(x,y,'m')
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
n = range(9)
t = np.linspace(-np.pi * 6, np.pi * 6, 1000)
for i in n:
    plt.subplot(3,3,i+1)
    x = t**i * np.cos(t)**3
    y = t**i * np.sin(t)**3

    plt.plot(x, y, 'k')
    plt.axis('square')
    plt.axis('off')
plt.show()
```

```python
k = np.linspace(0, 4, 9)
t = np.linspace(0, np.pi * 9, 1000)
for i in range(0, len(k)):
    plt.subplot(3, 3, int(i)+1)
    x = np.cos(k[int(i)]*t) * np.cos(t)
    y = np.cos(k[int(i)]*t) * np.sin(t)

    plt.plot(x, y, 'm')
    plt.axis('square')
    plt.axis('off')
plt.show()
```

```python
n = 150
t = np.linspace(0, np.pi * 200, 12304)
k = np.linspace(-10, -2, n)

for i in range(n):
    x = np.cos(t) * np.exp(t/k[i])
    y = np.sin(t) * np.exp(t/k[i])
    plt.plot(x, y, 'r',color=[i/n,0,i/n])
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
n = 500
rs = np.linspace(1, 4, 1000)
x = .5 + np.zeros(n)
endcap = np.arange(round(n*.9), n)
for ri in range(len(rs)):
    for ni in range(n-1):
        x[ni+1] = rs[ri]*x[ni]*(1-x[ni]) + .1*np.sin(6*x[ni])**2

    u = np.unique(x[endcap])
    r = rs[ri] * np.ones(len(u))
    plt.plot(r, u, 'k.',markersize=0.3)


# plt.axis('square')
# plt.axis('off')
plt.show()
```
