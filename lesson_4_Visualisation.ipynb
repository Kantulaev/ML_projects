```python
import sympy as sym
from IPython.display import display
from IPython.core.display import Math
import numpy as np
import matplotlib.pyplot as plt
from sympy.abc import x,y

sym.init_printing()
```

```python
y = x**2-3*x


for i in range(-10,12):
    plt.plot(i,y.subs(x,i),'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %y)
    
plt.show()
```
