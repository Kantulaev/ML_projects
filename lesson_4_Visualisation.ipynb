```python
import sympy as sym
from IPython.display import display
from IPython.core.display import Math
import numpy as np
import matplotlib.pyplot as plt
from sympy.abc import x
from matplotlib.patches import Polygon

sym.init_printing()
```

```python
y = x**2-3*x


for i in range(-10,12):
    plt.plot(i,y.subs(x,i),'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %y)
    
plt.show()
```

```python
y = sym.sqrt(abs(x))
xList = list(range(-20, 21))

for xI in xList:
    plt.plot([0,  xI], [0, y.subs(x, xI)],linewidth=3)

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot([0, 2], [0, 0],linewidth=3,color='g')
plt.plot([2, 2], [0, 2],linewidth=3,color='r')
plt.plot([2, 0], [2, 2],linewidth=3,color='k')
plt.plot([0, 0], [2, 0],linewidth=3,color='y')
plt.axis('square')
plt.axis([-2.5,4.5,-3,5])

plt.show()
```

```python
x = [-10, 10]
m = [0.7, -1.25]
b = [-2, 0.75]

for i in range(len(m)):
    y = m[i]*np.array(x) + b[i]
    plt.plot(x, y, label='y=%sx+%s' % (m[i], b[i]))

plt.axis('square')
plt.grid()
plt.legend()

plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphik')

plt.xlim(-5, 5)
plt.ylim(-5, 5)
axis = plt.gca()
plt.plot([0, 0], axis.get_xlim(), ('k--'))
plt.plot(axis.get_ylim(), [0, 0], ('k--'))

plt.show()
```

```python
x = np.linspace(-4, 4, 100)

for i in range(-1, 4):
    y = x**i
    plt.plot(x, y, ('-'), label='$y=x^{%s}$' % i, linewidth=4)

plt.legend()
plt.xlabel('x')
plt.ylabel('y')

plt.xlim(-4, 4)
plt.ylim(-20, 20)

plt.savefig('graphik.png')

plt.show()
```

```python
import sympy.plotting.plot as symplot
x, a = sym.symbols('x,a')
y = a/(x**2-a)
p = symplot(y.subs(a, 1), (x, -5, 5), show=False)

for i in range(2, 5):
    p.extend(symplot(y.subs(a, i), (x, -5, 5), show=False))
    p[i-1].label = '$%s$' % sym.latex(y.subs(a, i))
    p[i-1].line_color = list(np.random.rand(3))

p.ylim = [-10, 10]
p.legend = True
p.show()
```

```python
matr = np.zeros((8, 8))

for i in range(0, np.shape(matr)[0]):
    for j in range(0, np.shape(matr)[1]):
        matr[i, j] = (-1)**(i+j)

plt.set_cmap('gray')
plt.tick_params(labelleft=False, labelbottom=False)
plt.imshow(matr)
```

```python
x = np.linspace(-2,2,101)
y = -x**2
matrA = np.vstack((x,y)).T

matrB = np.array([[-0.5, -4], [0.5, -4], [0.5, -2.5], [-0.5, -2.5]])

pA = Polygon(matrA, facecolor='m', alpha=1)
pB = Polygon(matrB, facecolor='k', alpha=1)

fig, ax = plt.subplots()
ax.add_patch(pA)
ax.add_patch(pB)

ax.set_xlim([-2.5, 2.5])
ax.set_ylim([-4.5, 0.5])
plt.axis('off')
plt.show()
```
