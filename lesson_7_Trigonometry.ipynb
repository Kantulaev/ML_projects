```python
def main():

    angleInput = eval(input('What angle to convert?'))
    chooseInput = input('Which unit did you input(radian or degree)?')
    newAngle = 0
    plt.grid()
    plt.axis([-1, 1, -1, 1])
    plt.axis('square')
    plt.plot([0, 1], [0, 0], 'r')
    if chooseInput == 'radian':
        newAngle = np.rad2deg(angleInput)
        plt.plot([0, np.cos(newAngle)], 
                 [0, np.sin(newAngle)])
        plt.show()

    elif chooseInput == 'degree':
        newAngle = np.deg2rad(angleInput)
        plt.plot([0, np.cos(newAngle)],
                 [0, np.sin(newAngle)])
        plt.show()
    else:
        raise ValueError('Try again with correct input')


main()
```

```python
z = complex(2, 3)
lim = np.max((np.real(z), np.imag(z)))+1

plt.plot([0, np.real(z)], [0, np.imag(z)])
plt.plot([np.real(z), np.real(z)], [np.imag(z), 0], 'y--')
plt.plot(np.real(z), np.imag(z), 'yo', linewidth=5)
plt.text(lim-.5, 0+.2, 're', fontsize=16)
plt.text(0-.8, lim-.5, 'im', fontsize=16)

plt.plot([-lim, lim], [0, 0], 'k--')
plt.plot([0, 0], [-lim, lim], 'k--')
plt.axis([-lim, lim, -lim, lim])
plt.axis('square')
plt.show()

print((np.real(z)/np.imag(z))*100)
print(np.arctan2(np.imag(z),np.real(z)))
```

```python
th = np.linspace(0, (np.pi*11/6), 100)
lim = 1

Fth = np.tan(th)

# dot black
plt.plot(np.cos(th[-1]), np.sin(th[-1]), 'ko', linewidth=3)

# circle up to dot
plt.plot(np.cos(th), np.sin(th), 'k', linewidth=3)

# black thin
plt.plot([0, np.cos(th[-1])], [0, np.sin(th[-1])], 'k', alpha=.3)

# red and light red
plt.plot([0, 0], [0, np.sin(th[-1])], 'r', linewidth=3, label='cos part')
plt.plot([0, np.cos(th[-1])], [np.sin(th[-1]), np.sin(th[-1])], 'r', alpha=.3)

# blue and light blue
plt.plot([0, np.cos(th[-1])], [0, 0], 'c', linewidth=3, label='sin part')
plt.plot([np.cos(th[-1]), np.cos(th[-1])], [0, np.sin(th[-1])], 'c', alpha=.3)

# full circle
plt.plot(np.cos(np.linspace(0, np.pi*2, 1000)),
         np.sin(np.linspace(0, np.pi*2, 1000)), 'k', alpha=.3)

plt.plot([-lim, lim], [0, 0], 'k--', alpha=.2)
plt.plot([0, 0], [-lim, lim], 'k--', alpha=.2)
plt.axis([-lim, lim, -lim, lim])
plt.axis('square')
plt.legend(loc='lower left')
plt.title('$\\theta = \\frac{7\\pi}{6}$')
plt.show()
```
